```python
import numpy as np
import matplotlib.pyplot as plt

# Creating a random 4x4 array from an exponential distribution
array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print(array_4x4)
```

```python
# Creating random arrays
exp_array = np.random.exponential(scale=1.0, size=100000)
uniform_array = np.random.rand(100000)
normal_array = np.random.randn(100000)

# Plotting the histograms
plt.hist(exp_array, density=True, bins=100, histtype="step", color="green", label="exponential")
plt.hist(uniform_array, density=True, bins=100, histtype="step", color="blue", label="uniform")
plt.hist(normal_array, density=True, bins=100, histtype="step", color="red", label="normal")




# Setting the plot limits and labels
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Creating a grid of X and Y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plotting the 3D surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Setting the labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D plot of Z = X^2 + Y^2')
plt.show()
```
